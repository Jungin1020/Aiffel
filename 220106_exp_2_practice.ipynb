```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
digits=load_digits()
print(dir(digits))
```

```python
digit_data = digits.data
print(digit_data.shape)
digits.data[0]
```

```python
digits.target_names
```

```python
import pandas as pd

digits_df = pd.DataFrame(data = digit_data, columns = digits.feature_names)
digits_df
```

```python
digits_df['label'] = digits.target
digits_df
```

```python
digits_label= digits.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(digit_data,digits_label, test_size=0.2, random_state= 15 )
print(len(X_train), len(X_test))
```

```python
X_train.shape, y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


decision_tree = DecisionTreeClassifier(random_state=30)

decision_tree.fit(X_train,y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = DecisionTreeClassifier(random_state=30)

random_forest.fit(X_train,y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn import svm

svm_model = svm.SVC()

svm_model.fit(X_train,y_train)
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import SGDClassifier

sgd_model = SGDClassifier()

sgd_model.fit(X_train,y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression()

logistic_model.fit(X_train,y_train)
y_pred = logistic_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
from sklearn.datasets import load_wine

wine = load_wine()
wine_data = wine.data
print(wine_data.shape)
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
wine_label = wine.target
wine_df = pd.DataFrame(data = wine_data,columns=wine.feature_names)
wine_df['target']=wine.target

wine_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(wine_data,wine_label,test_size=0.2,random_state=15)
x_train.shape,x_test.shape, y_train.shape, y_test.shape
```

```python
decision_tree = DecisionTreeClassifier(random_state=30)

decision_tree.fit(x_train,y_train)
y_pred = decision_tree.predict(x_test)

print('의사결정나무')
print(classification_report(y_test, y_pred))
print('------------------------------------------------------')

random_forest = DecisionTreeClassifier(random_state=30)

random_forest.fit(x_train,y_train)
y_pred = random_forest.predict(x_test)

print('랜덤포레스트')
print(classification_report(y_test, y_pred))
print('------------------------------------------------------')

svm_model = svm.SVC()

svm_model.fit(x_train,y_train)
y_pred = svm_model.predict(x_test)

print('svm')
print(classification_report(y_test, y_pred))
print('------------------------------------------------------')

sgd_model = SGDClassifier()

sgd_model.fit(x_train,y_train)
y_pred = sgd_model.predict(x_test)

print('sdg')
print(classification_report(y_test, y_pred, zero_division = 0))
print('------------------------------------------------------')

logistic_model = LogisticRegression()

logistic_model.fit(x_train,y_train)
y_pred = logistic_model.predict(x_test)

print('로지스틱')
print(classification_report(y_test, y_pred,))

```
